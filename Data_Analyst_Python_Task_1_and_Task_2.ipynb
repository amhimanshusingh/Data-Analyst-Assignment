Question 1: Generate Car Matrix

```python
import pandas as pd

def generate_car_matrix():
    # Read the dataset-1.csv into a DataFrame
    df = pd.read_csv('dataset-1.csv')

    # Create a pivot table with id_1 as index, id_2 as columns, and car as values
    car_matrix = df.pivot_table(index='id_1', columns='id_2', values='car', fill_value=0)

    # Set diagonal values to 0
    for i in car_matrix.index:
        car_matrix.at[i, i] = 0

    return car_matrix

# Example usage
result_matrix = generate_car_matrix()
print(result_matrix)
```

Question 2: Car Type Count Calculation

```python
import pandas as pd

def get_type_count(df):
    # Add a new categorical column 'car_type' based on the values of the 'car' column
    df['car_type'] = pd.cut(df['car'], bins=[float('-inf'), 15, 25, float('inf')],
                            labels=['low', 'medium', 'high'], right=False)

    # Calculate the count of occurrences for each car_type category
    type_count = df['car_type'].value_counts().to_dict()

    # Sort the dictionary alphabetically based on keys
    type_count = dict(sorted(type_count.items()))

    return type_count

# Example usage
df = pd.read_csv('dataset-1.csv')
result_type_count = get_type_count(df)
print(result_type_count)
```

Question 3: Bus Count Index Retrieval

```python
import pandas as pd

def get_bus_indexes(df: pd.DataFrame) -> list:
    """
    Returns the indexes where the 'bus' values are greater than twice the mean.

    Args:
        df (pandas.DataFrame): Input DataFrame.

    Returns:
        list: List of indexes where 'bus' values exceed twice the mean.
    """
    # Calculate the mean value of the 'bus' column
    bus_mean = df['bus'].mean()

    # Identify indexes where 'bus' values are greater than twice the mean
    bus_indexes = df[df['bus'] > 2 * bus_mean].index.tolist()

    # Sort the list in ascending order
    bus_indexes.sort()

    return bus_indexes

# Example usage:
# Assuming df is the DataFrame read from 'dataset-1.csv'
# bus_indexes_result = get_bus_indexes(df)
```

### Question 4 Route Filtering

```python
import pandas as pd

def filter_routes(df: pd.DataFrame) -> list:
    """
    Filters and returns routes with average 'truck' values greater than 7.

    Args:
        df (pandas.DataFrame): Input DataFrame.

    Returns:
        list: List of route names with average 'truck' values greater than 7.
    """
    # Group by 'route' and calculate the average 'truck' values
    route_avg_truck = df.groupby('route')['truck'].mean()

    # Filter routes where the average 'truck' values are greater than 7
    selected_routes = route_avg_truck[route_avg_truck > 7].index.tolist()

    # Sort the list of routes in ascending order
    selected_routes.sort()

    return selected_routes

# Example usage:
# Assuming df is the DataFrame read from 'dataset-1.csv'
# filtered_routes_result = filter_routes(df)
```

### Question 6: Time Check

```python
import pandas as pd

def check_timestamp_completeness(df):
    # Combine date and time columns to create a datetime column
    df['start_datetime'] = pd.to_datetime(df['startDay'] + ' ' + df['startTime'])
    df['end_datetime'] = pd.to_datetime(df['endDay'] + ' ' + df['endTime'])

    # Check if timestamps cover a full 24-hour period
    hour_range_check = ((df['start_datetime'].dt.hour.min() == 0) and
                        (df['end_datetime'].dt.hour.max() == 23))

    # Check if timestamps span all 7 days of the week
    day_of_week_check = df['start_datetime'].dt.dayofweek.nunique() == 7

    # Group by (id, id_2) and check completeness for each group
    completeness_check = df.groupby(['id', 'id_2']).apply(lambda group: hour_range_check and day_of_week_check).astype(bool)

    return completeness_check

# Example usage
# Adjust the file path accordingly
df = pd.read_csv('dataset-2.csv')
result_series = check_timestamp_completeness(df)
print(result_series)
```

## Question 1: Distance Matrix Calculation

```python
import pandas as pd

def calculate_distance_matrix('/content/dataset-3.csv'):
    # Read the dataset-3.csv into a DataFrame
    df = pd.read_csv('/content/dataset-3.csv')

    # Create a DataFrame with IDs as index and columns
    distance_matrix = pd.pivot_table(df, values='Distance', index='ID1', columns='ID2', aggfunc='sum', fill_value=0)

    # Ensure the matrix is symmetric
    distance_matrix = distance_matrix + distance_matrix.T

    # Set diagonal values to 0
    distance_matrix.values[[range(distance_matrix.shape[0])]*2] = 0

    return distance_matrix

# Example usage
# Adjust the file path accordingly
distance_matrix = calculate_distance_matrix('dataset-3.csv')
print(distance_matrix)
```

### Question 2: Unroll Distance Matrix

```python
import pandas as pd

def unroll_distance_matrix(distance_matrix):
    # Reset the index to have 'ID1' and 'ID2' as columns
    distance_matrix_reset = distance_matrix.reset_index()

    # Melt the DataFrame to convert it to long format
    melted_df = pd.melt(distance_matrix_reset, id_vars='ID1', var_name='ID2', value_name='distance')

    # Exclude rows where id_start is equal to id_end
    unrolled_df = melted_df[melted_df['ID1'] != melted_df['ID2']].reset_index(drop=True)

    return unrolled_df

# Example usage
# Assuming distance_matrix is the DataFrame obtained from Question 1
# Adjust the DataFrame name based on your actual DataFrame
unrolled_distance_df = unroll_distance_matrix(distance_matrix)
print(unrolled_distance_df)
```

Question 3: Finding IDs within Percentage Threshold

```python
import pandas as pd

def find_ids_within_ten_percentage_threshold(df, reference_value):
    # Filter the DataFrame based on the reference value
    reference_df = df[df['id_start'] == reference_value]

    # Calculate the average distance for the reference value
    reference_avg_distance = reference_df['distance'].mean()

    # Calculate the threshold values
    lower_threshold = reference_avg_distance * 0.9
    upper_threshold = reference_avg_distance * 1.1

    # Filter the DataFrame based on the threshold values
    filtered_df = df[(df['distance'] >= lower_threshold) & (df['distance'] <= upper_threshold)]

    # Get the unique values from the 'id_start' column and sort them
    result_ids = sorted(filtered_df['id_start'].unique())

    return result_ids

# Example usage
# Assuming unrolled_distance_df is the DataFrame obtained from the previous question
# Adjust the DataFrame name based on your actual DataFrame
reference_value = 123  # Replace with the desired reference value
result_ids = find_ids_within_ten_percentage_threshold(unrolled_distance_df, reference_value)
print(result_ids)
```

### Question 4: Calculate Toll Rate

```python
import pandas as pd

def calculate_toll_rate(df):
    # Define rate coefficients for each vehicle type
    rate_coefficients = {'moto': 0.8, 'car': 1.2, 'rv': 1.5, 'bus': 2.2, 'truck': 3.6}

    # Create new columns for each vehicle type
    for vehicle_type, rate_coefficient in rate_coefficients.items():
        df[vehicle_type] = df['distance'] * rate_coefficient

    return df

# Example usage
# Assuming unrolled_distance_df is the DataFrame obtained from the previous question
# Adjust the DataFrame name based on your actual DataFrame
result_df = calculate_toll_rate(unrolled_distance_df)
print(result_df)
```

Question 5: Calculate Time-Based Toll Rates

```python
import pandas as pd
from datetime import time

def calculate_time_based_toll_rates(df):
    # Define time ranges and discount factors
    time_ranges = [(time(0, 0, 0), time(10, 0, 0)),
                   (time(10, 0, 1), time(18, 0, 0)),
                   (time(18, 0, 1), time(23, 59, 59))]

    discount_factors_weekdays = [0.8, 1.2, 0.8]
    discount_factor_weekends = 0.7

    # Create new columns for start_day, start_time, end_day, and end_time
    df['start_day'] = df['start_datetime'].dt.day_name()
    df['start_time'] = df['start_datetime'].dt.time
    df['end_day'] = df['end_datetime'].dt.day_name()
    df['end_time'] = df['end_datetime'].dt.time

    # Apply discount factors based on time ranges
    for time_range, discount_factor in zip(time_ranges, discount_factors_weekdays):
        mask = (df['start_time'] >= time_range[0]) & (df['start_time'] <= time_range[1])
        df.loc[mask, ['moto', 'car', 'rv', 'bus', 'truck']] *= discount_factor

    # Apply constant discount factor for weekends
    weekend_mask = (df['start_datetime'].dt.dayofweek >= 5)  # 5 and 6 represent Saturday and Sunday
    df.loc[weekend_mask, ['moto', 'car', 'rv', 'bus', 'truck']] *= discount_factor_weekends

    return df

# Example usage
# Assuming result_df is the DataFrame obtained from Question 4
# Adjust the DataFrame name based on your actual DataFrame
result_with_time_based_tolls = calculate_time_based_toll_rates(result_df)
print(result_with_time_based_tolls)
```
